```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df_training = pd.read_csv('lucas_organic_carbon/training_test/lucas_organic_carbon_training_and_test_data.csv')
df_training
```

```python
pca = PCA(n_components=100) 
df_compressed = pca.fit_transform(df_training)
```

```python
df_compressed = pd.DataFrame(df_compressed)
df_compressed
```

```python
df_compressed.to_csv('lucas_organic_carbon/training_test/compressed_data.csv', index=False)
```
